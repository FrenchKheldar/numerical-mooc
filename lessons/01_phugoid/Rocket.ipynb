```python
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
m_s = 50. # Mass of rocket shell in kg
g = 9.81 # Gravity in m/s^2
rho = 1.091 # air density in kg/m3
r = 0.5 # rocket radius in m
A = numpy.pi * r**2
print A
v_e = 325. # Exhaust speed in m/s
C_D = 0.15 # Drag coefficient
m_p0 = 100. # Initial mass of rocket propellant
```

```python
def burn_rate(t):
    """ Function that defines the rocket burning rate
    
    Parameters
    ----------
    t : float
        time in seconds
    """
    if t < 5.0:
        return(20)
    else:
        return(0)
```

```python
def f(u,t):
    """Returns the right-hand side of the rocket system of equations.
    
    Parameters
    ----------
    u : array of float
        array containing the solution at time n.
    t : float
        time in seconds
        
        
    Returns
    -------
    dudt : array of float
        array containing the RHS given u.
    """
    
    v = u[0]
    y = u[1]
    m = u[2]
    return numpy.array([-g + burn_rate(t) * v_e / (m_s + m) - 0.5 * rho * v *numpy.abs(v) * C_D * A / (m_s+m),
                      v,
                      -burn_rate(t)])
```

```python
def euler_step(u, f, dt, t):
    """Returns the solution at the next time-step using Euler's method.
    
    Parameters
    ----------
    u : array of float
        solution at the previous time-step.
    f : function
        function to compute the right hand-side of the system of equation.
    dt : float
        time-increment.
    
    Returns
    -------
    u_n_plus_1 : array of float
        approximate solution at the next time step.
    """
    
    return u + dt * f(u,t)
```

```python
dt = 0.1 # Timestep in s
x0 = 0.0 # Initial rocket location
y0 = 0.0 # Initial rocket location
v0 = 0.0 # Initial rocket vertical speed
T = 100.0 # Maximum time
N = int(T/dt) + 1                  # number of time-steps
t = numpy.linspace(0.0, T, N)      # time discretization
```

```python
# initialize the array containing the solution for each time-step
u = numpy.empty((N, 3))
u[0] = numpy.array([v0, y0, m_p0])# fill 1st element with initial values
n=0
print t[n], u[n]
# time loop - Euler method
while u[n,1] >= 0 and n < N:
    u[n+1] = euler_step(u[n], f, dt, t[n])
    print t[n+1],u[n+1]
    n = n + 1
    
```

```python
# visualization of the path
plt.figure(figsize=(8,6))
plt.grid(True)
plt.xlabel(r't(s)', fontsize=18)
plt.ylabel(r'y(m)', fontsize=18)
plt.title('Rocket trajectory, flight time = %.2f' % t[n], fontsize=18)
plt.plot(t[:n+1],u[:n+1,1], 'k-', lw=2);
```
